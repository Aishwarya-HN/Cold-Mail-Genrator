```python
import chromadb
client=chromadb.Client()
collection=client.create_collection(name='my_collection')
```

```python
collection.add(
    documents=['This document is about New York','This document is about delhi'],
    ids=['id1','id2']
)
```

```python
all_docs=collection.get()
all_docs
```

```python
one_doc=collection.get(ids=['id1'])
one_doc
```

```python
results=collection.query(query_texts=['Query is about chhole bhature'],n_results=2)
results
```

```python
results=collection.query(query_texts=['Query is about Brooklyn bridge'],n_results=2)
results
```

```python
results=collection.query(query_texts=['Query is about beaches'],n_results=2)
results
```

```python
results=collection.query(query_texts=['Query is about big apples'],n_results=2)
results
```

```python
collection.delete(ids=all_docs['ids'])
collection.get()
```

```python
collection.add(
    documents=['This document is about New york','This document is about Delhi'],
    ids=['id1','id2'],
    metadatas=[
        {"url":"https://en.wikipedia.org/wiki/New_York_City"},
        {"url":"https://en.wikipedia.org/wiki/New_Delhi"}
    ]
)
```
